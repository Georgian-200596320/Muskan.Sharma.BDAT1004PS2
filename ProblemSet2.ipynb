# Question 3
write a class named Kangaroo as a subclass of Marsupial that inherits all the attributes of Marsupial and also:
a. extends the Marsupial __init__ constructor to take, as input, the coordinates x and y of the Kangaroo object,
b. supports method jump that takes number values dx and dy as input and moves the kangaroo by dx units along the x-axis and by dy units along the yaxis, and
c. overloads the __str__ operator so it behaves as shown below.

```python
class Marsupial:
    def __init__(self):
        self.pouch = []
    
    def put_in_pouch(self, item):
        self.pouch.append(item)
    
    def pouch_contents(self):
        return self.pouch

class Kangaroo(Marsupial):
    def __init__(self, x=0, y=0):
        super().__init__()  # Initialize the pouch using the superclass constructor
        self.x = x
        self.y = y
    
    def jump(self, dx, dy):
        self.x += dx
        self.y += dy
    
    def __str__(self):
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"

# Creating a Marsupial object
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
m.pouch_contents() # returns: ['doll', 'firetruck', 'kitten']

# Creating a Kangaroo object
k = Kangaroo(0, 0)
print(k)  # Output: I am a Kangaroo located at coordinates (0,0)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())  # Output: ['doll', 'firetruck', 'kitten']

# Performing jumps
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)  # Output: I am a Kangaroo located at coordinates (3,0)
```

# Question 4
Write function collatz() that takes a positive integer x as input and prints the Collatz sequence starting at x. A Collatz
sequence is obtained by repeatedly applying this rule to the previous number x in the sequence:
x = { 𝑥/2 𝑖𝑓 𝑥 𝑖𝑠 𝑒𝑣𝑒𝑛
     3𝑥 + 1 𝑖𝑓 𝑥 𝑖𝑠 𝑜𝑑𝑑
Your function should stop when the sequence gets to number 1. Your implementation must be recursive, without any loops.


```python
def collatz(x):
    if x == 1:
        print(x)
    else:
        print(x, end='\n')
        if x % 2 == 0:
            collatz(x // 2)
        else:
            collatz(3 * x + 1)
print("collatz of 1 is:")
collatz(1)
print("collatz of 10 is:")
collatz(10)
```

# Question 5
Write a recursive method binary() that takes a non-negative integer n and prints the binary representation of integer n.

```python
def binary(n):
    if n < 2:
        print(n, end='')
    else:
        binary(n // 2)
        print(n % 2, end='')

# usage:
binary(0)
print()   # Print newline for clarity
binary(1)
print()  # Print newline for clarity
binary(3)
print()  # Print newline for clarity
binary(9)
```

# Question 6
Implement a class named HeadingParser that can be used to parse an HTML document, and retrieve and print all the headings in the document. You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser. When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your implementation using w3c.html.

```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.headings = []
        self.current_heading = None
    
    def handle_starttag(self, tag, attrs):
        if tag.startswith('h') and tag[1:].isdigit():
            self.current_heading = int(tag[1:]) - 1  # Convert h1, h2, ... to indices 0, 1, ...
    
    def handle_endtag(self, tag):
        if self.current_heading is not None:
            self.current_heading = None
    
    def handle_data(self, data):
        if self.current_heading is not None:
            self.headings.append((self.current_heading, data.strip()))
    
    def print_headings(self):
        for heading in self.headings:
            indent = heading[0]  # Indentation is based on the heading level
            print('    ' * indent + heading[1])

# usage:
if __name__ == "__main__":
    with open('w3c.html', 'r') as f:
        html_content = f.read()
        
    parser = HeadingParser()
    parser.feed(html_content)
    parser.print_headings()
```

# Question 7
Implement recursive function webdir() that takes as input: a URL (as a string) and non-negative integers depth and indent. 
Your function should visit every web page reachable from the starting URL web page in depth clicks or less, and print 
each web page's URL. 

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

def webdir(url, depth, indent=0, visited=None):
    if visited is None:
        visited = set()
    
    # Print the URL with appropriate indentation
    print('    ' * indent + url)
    
    # Base case: when depth is 0, stop further recursion
    if depth == 0:
        return
    
    try:
        # Make a request to fetch the HTML content of the URL
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for bad responses
        
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find all anchor tags (links) in the page
        links = soup.find_all('a', href=True)
        
        for link in links:
            href = link['href']
            
            # Construct the absolute URL using urljoin to handle relative links
            absolute_url = urljoin(url, href)
            
            # Normalize the URL to avoid duplicates and external links
            parsed_url = urlparse(absolute_url)
            if parsed_url.scheme not in ('http', 'https'):
                continue  # Skip non-http(s) links
            
            # Check if the absolute URL is within the same domain and not visited yet
            if absolute_url not in visited:
                visited.add(absolute_url)
                # Recursive call to webdir for each link found
                webdir(absolute_url, depth - 1, indent + 1, visited)
            
    except requests.exceptions.RequestException as e:
        # Print any exceptions that occur during the request
        print('    ' * (indent + 1) + f"Error: {e}")

# usage:
if __name__ == "__main__":
    url = 'https://www.w3schools.com/'
    depth = 2 
    
    print(f"Website directory for URL: {url}\n")
    webdir(url, depth)
```

# Question 8
Write SQL queries on the below database table that return: 
a) All the temperature data.
b) All the cities, but without repetition.
c) All the records for India.
d) All the Fall records
e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters.
f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order.
g) The total annual rainfall for Cairo.
h) The total rainfall for each season.

```python
import sqlite3

# Create a connection to an in-memory SQLite database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create WeatherData table
cursor.execute('''
    CREATE TABLE WeatherData (
        City TEXT,
        Country TEXT,
        Season TEXT,
        Temperature REAL,
        Rainfall REAL
    )
''')

# Insert data into WeatherData table
weather_data = [
    ('Mumbai', 'India', 'Winter', 24.8, 5.9),
    ('Mumbai', 'India', 'Spring', 28.4, 16.2),
    ('Mumbai', 'India', 'Summer', 27.9, 1549.4),
    ('Mumbai', 'India', 'Fall', 27.6, 346.0),
    ('London', 'United Kingdom', 'Winter', 4.2, 207.7),
    ('London', 'United Kingdom', 'Spring', 8.3, 169.6),
    ('London', 'United Kingdom', 'Summer', 15.7, 157.0),
    ('London', 'United Kingdom', 'Fall', 10.4, 218.5),
    ('Cairo', 'Egypt', 'Winter', 13.6, 16.5),
    ('Cairo', 'Egypt', 'Spring', 20.7, 6.5),
    ('Cairo', 'Egypt', 'Summer', 27.7, 0.1),
    ('Cairo', 'Egypt', 'Fall', 22.2, 4.5)
]

cursor.executemany('''
    INSERT INTO WeatherData (City, Country, Season, Temperature, Rainfall)VALUES (?, ?, ?, ?, ?)''', weather_data)

# Commit the changes
conn.commit()

# Queries
queries = {
    'a': 'SELECT Temperature FROM WeatherData;',
    'b': 'SELECT DISTINCT City FROM WeatherData;',
    'c': 'SELECT * FROM WeatherData WHERE Country = "India";',
    'd': 'SELECT * FROM WeatherData WHERE Season = "Fall";',
    'e': '''
        SELECT City, Country, Season
        FROM WeatherData
        GROUP BY City, Country, Season
        HAVING AVG(Rainfall) BETWEEN 200 AND 400;
    ''',
    'f': '''
        SELECT City, Country
        FROM WeatherData
        WHERE Season = "Fall"
        GROUP BY City, Country
        HAVING AVG(Temperature) > 20
        ORDER BY AVG(Temperature) ASC;
    ''',
    'g': 'SELECT SUM(Rainfall) AS TotalRainfall FROM WeatherData WHERE City = "Cairo";',
    'h': 'SELECT Season, SUM(Rainfall) AS TotalRainfall FROM WeatherData GROUP BY Season;'
}

# Execute each query and print results
for query_key, query_sql in queries.items():
    cursor.execute(query_sql)
    results = cursor.fetchall()
    print(f"\nQuery ({query_key}):")
    for result in results:
        print(result)

# Close the connection
conn.close()
```

# Question 9
Suppose list words is defined as follows:
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']
Write list comprehension expressions that use list words and generate the following lists:

```python
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

# part a
upper_case_words = [word.upper() for word in words]

# part b
lower_case_words = [word.lower() for word in words]

# part c 
word_lengths = [len(word) for word in words]

# part d
word_info = [[word.upper(), word.lower(), len(word)] for word in words]

# part e
words_4_or_more = [word for word in words if len(word) >= 4]

# Printing the results
print("a) ", upper_case_words)
print("b) ", lower_case_words)
print("c) ", word_lengths)
print("d) ", word_info)
print("e) ", words_4_or_more)
```
